```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('historical_automobile_sales.csv')
```

```python
df.head()
```

```python
%matplotlib inline
```

```python
df.dtypes
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
Year_sales=df.groupby('Year')['Automobile_Sales'].sum()
```

```python
Year_sales.plot(kind='line',marker='o',figsize=(10,6))
plt.title("Automobile_Sales By Year")
plt.xlabel('Year')
plt.ylabel('Total Automobile_Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.columns
```

```python
Car=df.groupby(['Year','Vehicle_Type'])['Automobile_Sales'].sum()
```

```python
Car.plot(kind='line', marker='o', figsize=(20,10), color='Orange')
plt.title(" Sales Trend By Vehicle ")
plt.xlabel('Car and Years')
plt.ylabel(' Car Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.head(10)
```

```python
grouped_df = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data=grouped_df,
    x='Vehicle_Type',
    y='Automobile_Sales',
    hue='Recession'
)
plt.title('Recession vs  Non-Recession')
plt.xlabel('Vehicle Type')
plt.ylabel('Average sales')
plt.tight_layout()
plt.show()
```

```python
avg_sales = df.groupby(['Vehicle_Type', 'Recession'])['Automobile_Sales'].mean()
avg_sales.plot(kind='barh')

plt.title('Vehicle Type (Recession vs Non-Recession)')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Sales')
plt.legend(['Non-Recession','Recession'], title='Period')
plt.tight_layout()
plt.show()
```
